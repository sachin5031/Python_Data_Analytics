```python
start_date='2009-01-01'
end_date='2018-01-01'
data=(df.index >=start_date)&(df.index <end_date)
df=df.iloc[data]
```

```python
df.plot(figsize=(10,4))
```

```python
quarter=df.rolling(window=12).mean()
```
