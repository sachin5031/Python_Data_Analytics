```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create three variables for seaborn plotting
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 12]
z = ['yes', 'no', 'yes', 'no', 'yes']
```

```python
sns.displot(x=x,y=y)
```

```python
sns.displot(y)
```

```python
sns.displot(y, bins=2)
```

```python
sns.displot(y, binwidth=2)
```

```python
sns.displot(x=x, hue=z)
```

```python
sns.jointplot(x=x,y=y)
```

```python
sns.displot(x=y,y=x)
```

```python
sns.displot(x=y,y=x, hue=z)
```

```python
sns.displot(x=y,y=x, bins=3)
```

```python
sns.displot(x=y,y=x, binwidth=3)
plt.xlabel(" X Values")
plt.ylabel(" Y Values")
```

```python
sns.relplot(x=x,y=y, hue=z)
```

```python
sns.relplot(x=x,y=y, kind='line',hue=z)
```

```python
sns.relplot(x=x,y=y, kind='line')
```

```python
data={'a':x,'b':y,'c':z}
df=pd.DataFrame(data)
```

```python
sns.catplot(data=df,x=z,y=x)
```

```python
sns.catplot(data=df,x=z,y=x, kind='bar')
```

```python
sns.catplot(data=df,x=z,y=x, kind='violin', color='r')
```

```python
sns.catplot(data=df,x=z,y=y, kind='box')
```
