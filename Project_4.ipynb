```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Course+Data+Set.csv',index_col='DATE',parse_dates=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.plot(figsize=(10,4))
```

```python
start_date='2007-01-01'
end_date='2018-01-01'
data=(df.index>=start_date)&(df.index<end_date)
df=df.iloc[data]
```

```python
print(df.head())
print(df.tail())
```

```python
df.plot(figsize=(10,4))
```

```python
yearly=df.rolling(window=12).mean()
```

```python
yearly.plot(figsize=(10,4))
plt.grid()
plt.show()
```
