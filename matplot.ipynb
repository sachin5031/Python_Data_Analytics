```python
import matplotlib.pyplot as plt
```

```python
a = [1, 2, 3, 4, 5,6,7]
b = [2, 16, 46, 38, 10,25,58]
```

```python
plt.plot(a,b)
```

```python
plt.bar(a,b)
```

```python
plt.scatter(a,b)
```

```python
plt.stem(a,b)
```

```python
plt.step(a,b)
```

```python
plt.stem(a,b)
plt.xlabel("A Values")
plt.ylabel('B values')
plt.title('Plotting with Integers')
```

```python
plt.plot(a,b, color='r',marker='*')
plt.xlabel("A Values")
plt.ylabel('B values')
plt.title('Plotting with Integers')
```

```python
import matplotlib.pyplot as plt

b = [1, 2, 3, 4, 5]  # x-axis
a = [10, 20, 25, 30, 40]  # y-axis

plt.plot(b, a)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')
plt.grid()
plt.show()
```
